```python
class ATM:
    
    def __init__(self):
        self.pin = int(input("Enter Your Pin")) 
        self.balance = int(input("Enter Your Initial Balance "))
        print(" ___________Your pin and balance initialize sucessfully_________________")
        
        
    def change_pin(self):
        old_pin=int(input("Enter Your Old pin "))
        if old_pin==self.pin:
            new_pin=int(input("Enter Your New pin "))
            self.pin=new_pin
            print("Sucessfully change the pin ")
        else:
            print(" wrong pin !!!!!! ")

    def enter_pin(self):
        counter=3
        flag=True
        while flag:
            Pin=int(input("Enter The pin Number "))
            if self.pin==Pin:
                print(f" Sucessfully login on {counter} attempt") 
                flag=False
                return True
            else:
                print(f"wrong pin you have {counter-1} times !!!\n")
                counter=counter-1
                if counter==0:
                    flag=False
                    print(" Your limit is more than 3 times ")
                    return False
                    
            


    def check_balance(self):
         result=self.enter_pin()
         if result:
             print(f"Your current balance is:Rs. {self.balance}")
         else:
             print(" sorry you entered the pin wrong")

    
             
    def deposit(self):
        result=self.enter_pin()
        if result:
            amount=int(input("Enter The amount to be deposit"))
            if amount > 0:
                self.balance = self.balance + amount
                print(f"${amount} deposited. Your new balance is: Rs. {self.balance}")
            else:
                print("Invalid deposit amount.")
        else:
            print("!!!!!!!!!!!! wrong Pin !!!!!!!!!!!!!!!")



    def withdraw(self):
        result=self.enter_pin()
        if result:
            amount=int(input("Enter The amount to be withdraw "))
            if amount > 0:
                if amount <= self.balance:
                    self.balance =self.balance - amount
                    print(f"${amount} withdrawn. Your new balance is: Rs. {self.balance}")
                else:
                    print("Insufficient funds.")
            else:
                print("Invalid withdrawal amount.")
        else:
            print("!!!!!!!!!!!! Wrong Pin Number   !!!!!!! ")

```

```python
a1=ATM()
```

```python
a1.check_balance()
```

```python
a1.change_pin()
```

```python
a1.deposit()
```

```python
a1.withdraw()
```
