```python
#marker , markersize
```

```python
plt.figure(figsize=(10,3)) #height and weidth respectively
plt.plot(df["index"],df["RG Sharma"],marker = 'd',markersize=10,color="red",linewidth=2,markerfacecolor="blue",linestyle = ":") #linestyle can be given in name or style      #marker d is dimond shape  = can be given any

plt.xticks(df["index"].tolist())

plt.xlabel("seasons") #x-axis 
plt.ylabel("runs")    #y-axis
plt.title("comprasion between rohit and kolhi ") # main title
plt.show()  #it remvoes the data shown before the figure
```

```python
#comparing more than 2
```

```python
plt.figure(figsize=(10,3)) #height and weidth respectively
plt.plot(df["index"],df["RG Sharma"],marker = 'd',markersize=4,color="red",label="Rohit")
plt.plot(df["index"],df["V Kohli"],marker = 'o',markersize=4,color = "blue",label="Kolhi")

plt.xticks(df["index"].tolist())

plt.xlabel("seasons") #x-axis 
plt.ylabel("runs")    #y-axis
plt.title("comprasion between rohit and kolhi ") # main title
plt.grid()  #grid appers on this
plt.legend(loc= "upper left") # legend function this shows the name in the plot and loc change the place of that

plt.savefig("rohit.png") #this saves in image formate (any)
plt.show()  #it remvoes the data shown before the figure
```
