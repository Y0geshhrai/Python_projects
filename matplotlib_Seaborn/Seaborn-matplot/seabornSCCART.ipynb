## subplots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
```

```python
df=pd.read_csv("tips.csv")
df
```

```python
fig, axes = plt.subplots(1,2 , figsize=(12,3))
sns.countplot(data=df,x="smoker",ax=axes[0],hue="sex")
axes[0].set_title("Histogram of Total Bill")
sns.scatterplot(data=df,x="total_bill",y="tip",ax=axes[1],hue="smoker")
axes[1].set_title("total bill vs tips")
```

```python

fig, axes = plt.subplots(2,2 , figsize=(12,6))

sns.countplot(data=df,x="smoker",ax=axes[0,0],hue="sex")
axes[0,0].set_title("Histogram of Total Bill")

sns.scatterplot(data=df,x="total_bill",y="tip",ax=axes[0,1],hue="smoker")
axes[0,1].set_title("total bill vs tips")

sns.histplot(data=df,x="total_bill",bins=[0,10,20,30,40,50,60],ax=axes[1,0])
sns.kdeplot(data=df,x="total_bill",ax=axes[1,1],hue="sex",fill=True)
axes[1,1].set_title("kde plots")

plt.tight_layout()
plt.savefig("figure_122.png")
plt.show()
```
