# Import The Libraries 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
# player_details_links : /html/body/section[3]/div/div/div[2]/div/div[1]/div/a[24]
#player_names: /html/body/section[3]/div/div/div[2]/div/div[1]/div/a[24]/div/div/div[1]/h4
#right_hand_or_left : /html/body/section[3]/div/div/div[2]/div/div[1]/div/a[24]/div/div/div[1]/p[2]
# image_url : /html/body/section[3]/div/div/div[2]/div/div[1]/div/a[16]/div/div/div[2]/div/img

# player_info= /html/body/section[2]/div/div/div[2]/div/div[1]
# player_info= /html/body/section[2]/div/div/div[2]/div/div[1]

# professional_info=/html/body/section[2]/div/div/div[2]/div/div[2]
```

```python
driver = webdriver.Chrome()
driver.get('https://cricketnepal.org.np/squads/nepal-mens-senior-national-team/')
```

# Extract The Links Of The datas 

```python
#driver.quit()
import numpy as np

image_urls=[]
links_urls=[]

time.sleep(3)

for i in range(1,25):
    try:
        link_element = driver.find_element(By.XPATH, f'/html/body/section[3]/div/div/div[2]/div/div[1]/div/a[{i}]')
        link_url = link_element.get_attribute('href')
        print(link_url)
        links_urls.append(link_url)
    except:
        links_urls.append(np.nan)


    try:
        image = driver.find_element(By.XPATH,f'/html/body/section[3]/div/div/div[2]/div/div[1]/div/a[{i}]/div/div/div[2]/div/img')
        image_url = image.get_attribute('src')
        image_urls.append(image_url)
        print('image url ',image_url)
        
    except:
        image_urls.append(np.nan)
        
```

```python
len(image_urls)
```

# Extract all The players Details 

```python
# links_url 

# player_info= /html/body/section[2]/div/div/div[2]/div/div[1]
# player_info= /html/body/section[2]/div/div/div[2]/div/div[1]

# professional_info=/html/body/section[2]/div/div/div[2]/div/div[2]

player_info=[]
profesional_info=[]
image_urls=[]

for i in links_urls:
    driver.get(i)
    time.sleep(6)

    # Extract the Image Urls 
    try:
        image = driver.find_element(By.XPATH,f'/html/body/section[2]/div/div/div[1]/div/img')
        image_url = image.get_attribute('src')
        image_urls.append(image_url)
        print('image url ',image_url)
        
    except:
        image_urls.append(np.nan)


    # Extract the Image Details Informations
    try:
        player = driver.find_element(By.XPATH, f'/html/body/section[2]/div/div/div[2]/div/div[1]').text
        player_info.append(player)
        
    except:
        player_info.append(np.nan)

    # Extract the professional Informations 
    try:
        pro_info = driver.find_element(By.XPATH, f'/html/body/section[2]/div/div/div[2]/div/div[2]').text
        profesional_info.append(pro_info)
        
    except:
        profesional_info.append(np.nan)
        
```
