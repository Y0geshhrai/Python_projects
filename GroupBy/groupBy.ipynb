```python
employee_df[employee_df["Department"]=="IT"]["Salary"].mean() #avg salary of it 
```

```python
employee_df[employee_df["Department"]=="HR"]["Salary"].mean() #avg salary of it 
```

```python
grouped = employee_df.groupby("Department")
```

```python
grouped["Salary"].mean()
```

```python
grouped["Salary"].max()
```

```python
grouped["Salary"].min()
```

```python
grouped["Salary"].count() # any head can give value while counting
```

```python
summary_df=grouped.agg({"Salary":["min","max","count","sum","mean"],"Age":{"mean","max","min"}})
```
